```python
import tensorflow as tf 
from sklearn.metrics import classification_report,confusion_matrix, f1_score, auc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import cv2 
import pickle
import seaborn as sns
```

```python
def plot_models(log_inception,log_resnet,log_densenet=None,logtitle="Model-Plot (Fold 1)",plotwhich="accuracy"):
  plt.figure(figsize=(8, 6))  
  acc1 = log_inception[plotwhich]  
  epochs1 = range(len(acc1))
  plt.plot(epochs1, acc1, c="blue", label="Inception")  
  
  acc2 = log_resnet[plotwhich]  
  epochs2 = range(len(acc2))  
  plt.plot(epochs2, acc2, c="red", label="Resnet")   

  acc3 = log_densenet[plotwhich]  
  epochs3 = range(len(acc3))  
  plt.plot(epochs3, acc3, c="green", label="Densenet")   

  plt.xlabel("epochs")
  plt.ylabel(plotwhich)
  plt.title(logtitle)
  plt.legend()

  plt.savefig("Testing-Plots\\"+logtitle+".png")
```

# Accuracies Plot

```python
log_d = pickle.load(open("Logs\densenet_history_1.pickle", "rb"))
log_i = pickle.load(open("Logs\inception_history_1.pickle", "rb"))
log_r = pickle.load(open("Logs\\resnet_history_1.pickle", "rb"))
plot_models(log_i, log_r, log_d, logtitle="Accuracy-Epoch Plot (Fold - 1)", plotwhich="accuracy")
```
